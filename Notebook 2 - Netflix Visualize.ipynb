# Netflix Data Visualization

Load data prepped in a separate Notebook.

```
import pandas as pd

# Load the data
netflix_data = pd.read_csv('What_We_Watched_A_Netflix_Engagement_Report_2023Jan-Jun.csv')

# Display the first few rows of the dataframe
netflix_data.head()
```

### Visualizations

A bar chart showing the top 10 TV shows/movies based on viewing hours.

```
import matplotlib.pyplot as plt

# Sort dataframe by 'Hours Viewed' in descending order and take the top 10
top10_viewed = netflix_data.sort_values('Hours Viewed', ascending=False).head(10)

# Plot
plt.figure(figsize=(10,6))
plt.barh(top10_viewed['Title'][::-1], top10_viewed['Hours Viewed'][::-1], color='skyblue')
plt.xlabel('Hours Viewed (in hundred millions)')
plt.title('Top 10 Viewed Shows/Movies')
plt.show()
```

A pie chart to illustrate the proportion of English vs. Foreign shows.

```
labels = netflix_data['English / Foreign'].unique()

sizes = netflix_data['English / Foreign'].value_counts()

# Craft pie chart
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Ensure pie is drawn as a circle
ax1.axis('equal')

plt.title('Proportion of English vs. Foreign Shows/Movies')
plt.show()
```

Proportionate availability of TV shows and Movies in our dataset.

```
tv_movie_counts = netflix_data['TV / Movie'].value_counts()

plt.figure(figsize=(8,6))
plt.bar(tv_movie_counts.index, tv_movie_counts.values, color=['orange', 'green'])
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Proportion of TV Shows vs. Movies')
plt.show()
```

A line chart to analyze if shows released recently receive more views. Note that the dataset contains views for Titles available to watch in the past 6 months only. Not every Title has a value for the Release Date field.

```
netflix_data['Release Date'] = pd.to_datetime(netflix_data['Release Date'])
netflix_data.sort_values('Release Date', inplace=True)

plt.figure(figsize=(10,6))
plt.plot_date(netflix_data['Release Date'], netflix_data['Hours Viewed'], linestyle='solid')
plt.gcf().autofmt_xdate()
plt.title('Release Date vs Viewing Hours')
plt.xlabel('Release Date')
plt.ylabel('Hours Viewed')
plt.tight_layout()
plt.show()
```

A pie chart to visualize the proportion of shows available globally.

```
labels = netflix_data['Available Globally?'].unique()

sizes = netflix_data['Available Globally?'].value_counts()

# Craft pie chart
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Ensure pie is drawn as a circle
ax1.axis('equal')

plt.title('Proportion of Shows Available Globally')
plt.show()
```
